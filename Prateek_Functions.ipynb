6.  Explain the concept of generators in Python and how they are defined?

def gen(n):
  count = 1
  while count<= n:

print(next(gen))

every next(gen) gets us the next value in the iteration

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
-> Map function maps a function to all the elements of an iterable. reduce() function reduces or condences the result into a single output. filter() function is typically used to filter out the elements based on the conditions
Eg:

```python
from functools import reduce
lst = [1,2,3,4]
#maps function to all items of the list
k = list(map(lambda x:x*2, lst))
print ("Map function output: ", k)

#multiplies all the number and generate only 1 output
l = reduce(lambda x,y:x*y, lst)
print ("Reduce function output: ", l)

#Filters even number
m = list(filter(lambda x:x%2==0, lst))
print ("Filter function output: ", m)
```

```python
#5  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

n = int(input())
fib = Fibonacci(n)


for num in fib:
    print(num, end=' ')
```

```python
#6  Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power(exp):
    for e in range(exp + 1):
        yield 2 ** e

exp = int(input())
for num in power(exp):
    print(num)
```

```python
#7  Implement a generator function that reads a file line by line and yields each line as a string.
def read_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line

#can test by adding file to google drive
file_path = input()
for line in read_file(file_path):
    print(line.strip())
```

```python
#8  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [(4, 10), (1, 5), (3, 2), (2, 8)]

# Sort using lambda: key is the second element (index 1)
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
#10 Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_str))

text = input("Enter a string: ")
result = remove(text)
print("String without vowels:", result)
```
